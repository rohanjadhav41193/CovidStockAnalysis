# Data preprocessing 
### removing unnecessary special characters to make data capable for ploting and analysis 
### conversion of Datatype into Date format
### checking null data 
### removing duplicate records which we get in web scraping 

```python
# importing data set which is in csv format to pandas dataframe (jordenFile)
jordenFile = pd.read_csv('UnstructuredCoreData.csv')
```

```python
# removing ',' from column so that it would be easy to change it in number format
jordenFile['Atleast1dose']=jordenFile['Atleast1dose'].str.replace(',','')
```

```python
# removing ',' from column so that it would be easy to change it in number format
jordenFile['FullVaccination']=jordenFile['FullVaccination'].str.replace(',','')
```

```python
# removing '%' from column to take it as integer so that we could use it in ploting and analysis 
jordenFile['DosePer']=jordenFile['DosePer'].str.replace('%','')
```

```python
# removing '%' from column to take it as integer so that we could use it in ploting and analysis
jordenFile['FullVaccPer']=jordenFile['FullVaccPer'].str.replace('%','')
```

```python
# for ploting and other actions we converted Date column to Date time format
jordenFile['Date']=pd.to_datetime(jordenFile['Date'])  
```

```python
# current columns of Dataframe 
jordenFile.columns
jordenFile.shape
```

```python
# we are segrigating Date column into three diffrent columns so that it could be helpful for analysis 
jordenFile['Date']=pd.to_datetime(jordenFile['Date'])
jordenFile['Year1']=jordenFile['Date'].dt.year
jordenFile['Month1']=jordenFile['Date'].dt.month
jordenFile['Date1']=jordenFile['Date'].dt.day
```

```python
# removing outliers its wrong data which we have received while scrapping 
indexNames =jordenFile[ jordenFile['Month1'] == 7].index
jordenFile.drop(indexNames , inplace=True)
```

```python
jordenFile.shape
```

```python
# for time being Year1,Month1,Date1 columns are not required so we are droping them so that our table size would be small 
jordenFile.drop(['Date1'],axis=1,inplace=True)
jordenFile.drop(['Month1'],axis=1,inplace=True)
jordenFile.drop(['Year1'],axis=1,inplace=True)
```

```python
jordenFile.head(5)
```

```python
# as we could see we have received some duplicate rows in datafarme so wre are removing those with below query 

JordenFinal=jordenFile[~jordenFile.duplicated(subset=['Country','Date','Atleast1dose','FullVaccination','DosePer','FullVaccPer'])]
JordenFinal
```

```python
# removed unnecessary column which was generated in data preprocessing 
JordenFinal.drop('Unnamed: 0',axis=1,inplace=True)
JordenFinal.shape
```

```python
# storing processed data on physical location

import os
if os.path.exists("ProcessedAndTransformedData.csv"):
     os.remove("ProcessedAndTransformedData.csv")
else:
    print("The file does not exist")

JordenFinal.to_csv('ProcessedAndTransformedData.csv',index=True)
```

# Transfering data in unstructured database MongoDB

```python
# we are using MongoDb as unstructure database for data storage ,we are creating database collection and storing data in MongoDB
# raw data will be first transfered to MongoDB

try:
    import pymongo
    from pymongo import MongoClient
    import pandas as pd
    import json
except Exception as e:
    print("Some Modules are Missing ")


class MongoDB(object):

    def __init__(self, dBName=None, collectionName=None):

        self.dBName = dBName
        self.collectionName = collectionName

        self.client = MongoClient("localhost", 27017, maxPoolSize=50)

        self.DB = self.client[self.dBName]
        self.collection = self.DB[self.collectionName]



    def InsertData(self, path=None):
        """

        :param path: Path os csv File
        :return: None
        """

        df = pd.read_csv(path)
        data = df.to_dict('records')

        self.collection.insert_many(data, ordered=False)
        print("All the Data has been Exported to Mongo DB Server .... ")

if __name__ == "__main__":
    mongodb = MongoDB(dBName = 'Dataset', collectionName='covidVac2304')
    mongodb.InsertData(path="ProcessedAndTransformedData.csv")
```

```python
# created function which would make connection with cloud database ,fetched data from dataframe which get data from local Postgres database.
# and stored data into table which is created on cloud platform.
def execute_values(connection,df, table):

    conn_string = "host={0} user={1} dbname={2} password={3} sslmode={4}".format(host, user, dbname, password, sslmode)
    azureConnection = psycopg2.connect(conn_string)
    print("Connection established")

    azureCursor = azureConnection.cursor()
   
    tuples = [tuple(x) for x in df.to_numpy()]
    
   
    cols = ','.join(list(df.columns))
    
    azureCursor = azureConnection.cursor()
   
    query  = "INSERT INTO %s(%s) VALUES %%s" % (table, cols)
    try:
        extras.execute_values(azureCursor, query, tuples)
        azureConnection.commit()
    except (Exception, psycopg2.DatabaseError) as error:
        print("Error: %s" % error)
        azureConnection.rollback()
        azureCursor.close()
        return 1
    print("execute_values() done")
    azureCursor.close()
```

```python
# execute function
execute_values(azureConnection, df ,"CovidVaccination")
```
