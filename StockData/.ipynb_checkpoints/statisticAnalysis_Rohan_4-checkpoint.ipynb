### Stacked bar-chart : Stockwise representation of Closed Value and Adj closed Value

```python
### Stacked bar chart
### Shows the Adj closed values against Actual Close values

europeanstockDF  = df[(df.stockName == "SAP") ]


europeanStockHighLowDF = europeanstockDF[['stockDate','openValue', 'closeValue']]

europeanStockHighLowDF['stockDate'] = pd.to_datetime(europeanStockHighLowDF['stockDate']).dt.normalize()

europeanStockHighLowDF1 = europeanStockHighLowDF.resample("M",on='stockDate').mean()


fig, ax = plt.subplots()

europeanStockHighLowDF1.plot(kind="bar", figsize=(16,5), ax=ax)

ax.set_xticklabels([x.strftime("%Y-%m") for x in europeanStockHighLowDF1.index], rotation=45)

plt.title("Open Value Vs Close Value for SAP stock")
plt.xlabel("Month")
plt.ylabel("Price of stock")
plt.show()
```

### Line charts : For stock how much the close values is been upscaled or downscaled 

```python
### Function created to plot the line chart for provides stock names
### this has one input parameter i.e stock list 
### pass the stock list it will plot the line chart for those 

def plot_stock_close(stockList):
    for u in stockList[:len(stockList)]:
        dates = dateAndCloseDF[(dateAndCloseDF["stockName"] == u)]["stockDate"]
        values = dateAndCloseDF[(dateAndCloseDF["stockName"] == u)]["closeValue"]
        plt.plot(dates.tolist(), values.tolist(),markersize=50)
    plt.legend(stockList, bbox_to_anchor=(1.05, 1), loc=2)
    plt.xlabel('Time (months)', fontsize=13)
    plt.ylabel('Price of stock)', fontsize=12)
   # plt.figure(figsize=(50,10))
    fig = plt.gcf()
    fig.set_size_inches(12,8)
    plt.show()
```

```python
## Calling the function by passing the list of EU based stocks

plot_stock_close(europeanStocksList)
```

```python
## Calling the function by passing 10 stock of EU based stocks but country level 

plot_stock_close(otherCountryWisestockList[:10])
```

### Top 10 gained stocks in COVID-19 time period

```python
### Below logic gives the top 10 stock whose close value gain more
### Creates the dataframe and plots the line chart for the same

symbols = df["stockName"].unique().tolist()
maxPerSymbol={}
for u in symbols:
    values = df[(df["stockName"] == u)]["closeValue"].tolist()
    maxPerSymbol[u] = max(values)

top10Closers = list(dict(sorted(maxPerSymbol.items(), 
                   key=lambda v:v[1], #sort by key -> v[0] | sort by value->v[1]
                   reverse=True)[:10]).keys())

top_10_mask = [i in top10Closers for i in df["stockName"].tolist()]
df_masked2 = df[top_10_mask]

df_masked2
```

```python
### Calling the function by passing the list of stock which are tope gained in COVID-19 time period

plot_stock_close(df_masked2.stockName.unique().tolist())
```

### Adding one more column named 'diff'- how much a stock values changed for that day

```python
### Introducing one more column that is diff which show by what amount the value changed for that day 

europeanStocksDF['diff'] = europeanStocksDF['closeValue'] - europeanStocksDF['openValue']

europeanStocksDF
```

```python
### this function created so that anyone can pass the list of stock this will extract the diff columns and plot the line chart

def plot_stock_close_for_diff(stockList):
    for u in stockList[:len(stockList)]:
        
        dates = europeanStocksDF[(dateAndCloseDF["stockName"] == u)]["stockDate"]
        values = europeanStocksDF[(dateAndCloseDF["stockName"] == u)]["diff"]
        plt.plot(dates.tolist(), values.tolist(),markersize=50)
        
    plt.legend(stockList, bbox_to_anchor=(1.05, 1), loc=2)
    plt.xlabel('Time (months)', fontsize=13)
    plt.ylabel('% Price of stock gain/loss )', fontsize=12)
    fig = plt.gcf()
    fig.set_size_inches(12,5)
    plt.show()
```

```python
### taking list of stock involved at EU level 
### Plotting the line charte for change for 5 initial stock

listTop10 = europeanStocksDF.stockName.unique().tolist()
for i in [0,1,2,3,4]:
    plot_stock_close_for_diff(listTop10[i:i+1])
```
